```python
c1 = ['Stock', 'Open', 'Volume', 'Close']
df = data[c1]
df
```

```python
# data cleaning

df.Volume = df.Volume.str.replace(",","")
df.Volume.astype("str")
df.Volume.replace({'' : 0, ' ' : 0, None : 0, "nan" : 0}, inplace= True)
df.Volume = df.Volume.astype("float")
print(df.Volume.dtype)
```

```python
df.Close = df.Close.str.replace(",","")
df.Close.astype("str")
df.Close.replace({'' : 0, ' ' : 0, None : 0, "nan" : 0}, inplace= True)
df.Close = df.Close.astype("float")
print(df.Close.dtype)
```

```python
df.Open = df.Open.str.replace(",","")
df.Open.astype("str")
df.Open.replace({'Dividend' : 0, 'Stock Split' : 0, '' : 0, ' ' : 0, None : 0, "nan" : 0}, inplace= True)
df.Open = df.Open.astype("float")
df.Open.replace({0 : df.Open.median()}, inplace=True)
print(df.Open.dtype)
```

```python
# removeing the outliers
Q1 = df.quantile(0.25)
Q3 = df.quantile(0.75)
IQR = Q3 - Q1
lower = (Q1 - 1.5 * IQR)
upper = (Q3 + 1.5 * IQR)
df = df[~((df < lower) & (df > upper)).any(axis=1)]
```

```python
df.dropna(inplace=True)
```
